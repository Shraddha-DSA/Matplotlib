#Scatter Plot

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,4,8,9,8,7]
plt.scatter(x,y)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,4,8,9,8,7]
colors=["r","y","g","b","r","g"]
s = [10,20,150,27,28,11,15]
plt.scatter(x,y,c=colors,sizes=s,alpha=0.2,marker="*",edgecolor = "g",linewidth=2) #alpha is used for transparency,marker is used for shape
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day")
plt.ylabel("No.")
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [2,4,8,9,8,7,3]
colors=[10,49,30,29,56,20,30]
s = [10,20,150,27,28,11,15]
plt.scatter(x,y,c=colors,sizes=s,cmap="viridis")
t=plt.colorbar()
t.set_label("ColorBar")
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day")
plt.ylabel("No.")
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [2,4,8,9,8,7,3]
y2 = [3,2,1,4,5,3,6]
colors=[10,49,30,29,56,20,30]
s = [10,20,150,27,28,11,15]
plt.scatter(x,y,c=colors,sizes=s,cmap="viridis")
plt.scatter(x,y2,c="r",sizes=s)
t=plt.colorbar()
t.set_label("ColorBar")
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day")
plt.ylabel("No.")
plt.show()
```

#Fill_Between

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])


plt.plot(x,y,color="r")

plt.fill_between(x,y,color="g",where=(x>=2)&(x<=4),alpha=0.4)

plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y,color="r")

plt.fill_between(x=[2,3],y1=2,y2=4)

plt.title("Python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
